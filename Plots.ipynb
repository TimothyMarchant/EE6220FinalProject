```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
```

```python
print(np.__version__)
print(matplotlib.__version__)
```

```python
#Emergency center outputs
#Bandwidth from a middlebox to the emergency center reported via iperf
IPerfNoSDNNoApplications=np.array([[35.7,36.6],[31.7,32.1],[30.8,31.6]])
IPerfNoSDNApplications=np.array([[35.1,35.6],[28.5,39.5],[34.5,35.0]])
IPerfSDNNoApplications=np.array([[30.9,39.4],[36.8,37.4],[35.2,35.9]])
IPerfSDNApplications=np.array([[13.1,13.8],[32.2,33.9],[37.7,42.4]])
```

```python
print(np.shape(IPerfNoSDNApplications))
```

```python
IperfavgNoSDNNoapp=np.average(IPerfNoSDNNoApplications,axis=0)
IperfavgNoSDNapp=np.average(IPerfNoSDNApplications,axis=0)
IperfavgSDNNoapp=np.average(IPerfSDNNoApplications,axis=0)
IperfavgSDNapp=np.average(IPerfSDNApplications,axis=0)
ArrayTransmit = np.array([IperfavgNoSDNNoapp[0],IperfavgNoSDNapp[0],IperfavgSDNNoapp[0],IperfavgSDNapp[0]])
ArrayResponse = np.array([IperfavgNoSDNNoapp[1],IperfavgNoSDNapp[1],IperfavgSDNNoapp[1],IperfavgSDNapp[1]])
Labels=["No SDN no Apps", "No SDN apps", "SDN no Apps", "SDN Apps"]
print(ArrayTransmit)
print(ArrayResponse)
```

```python
#Transmit bar graph
plt.bar(Labels,ArrayTransmit)
plt.xlabel("Type")
plt.ylabel("Goodput in Mb/s")
plt.title("Transmission iperf results, averaged; from Middleboxes to Emergency Center")
plt.show()
#Response bar graph
plt.bar(Labels,ArrayResponse)
plt.xlabel("Type")
plt.ylabel("Goodput in Mb/s")
plt.title("Response iperf results, averaged; from Middleboxes to Emergency Center")
plt.show()
```
